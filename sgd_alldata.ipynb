```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import SGDClassifier
from sklearn.preprocessing import StandardScaler
```

```
df = pd.read_csv('/content/drive/MyDrive/AMaszynowe/Reuters_exclude_usa.csv')
```

```
df = pd.read_csv('/content/drive/MyDrive/AMaszynowe/ReutersPY.csv')
```

```
df = df.iloc[:,:-1]
df.fillna(0)
```

```
df.labels = pd.factorize(df.labels)[0]

X = df.iloc[:,1:].values
y = df['labels'].values

np.unique(y)
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```
sgd_clf = SGDClassifier()
sgd_clf.fit(X_train, y_train)
```

```
y_preds = sgd_clf.predict(X_test)
```

```
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,y_preds))
```

```
from sklearn.metrics import precision_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
```

```
precision_score(y_test, y_preds,average='weighted')
```

```
accuracy_score(y_test, y_preds)
```

```
recall_score(y_test, y_preds, average='weighted')
```
